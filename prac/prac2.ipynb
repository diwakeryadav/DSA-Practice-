```python
from typing import List
class Solution:
    def twosum(self, nums: List[int], target: int) -> List[int]:
        for i in range(len(nums)):
            for j in range(i+1, len(nums)):
                if nums[i] + nums[j] == target:
                    return [i,j]
```

# 3. Remove Duplicates from sorted array
## Problem Statement 
#### A numerical array sorted in non-decreasing order (items are either greater than equal to, allowing duplicates) remove the duplicates so that the items are left with only unique values. the relative order should be the same
#### Consider the number of unique elements of nums to be k, to get accepted, you need to do the following things:

Change the array nums such that the first k elements of nums contain the unique elements in the order they were present in nums initially. The remaining elements of nums are not important as well as the size of nums.
Return the count of the unique elements in num

#### Input: nums = [1,1,2]
#### Output: 2, nums = [1,2,_]

#### Input: nums = [0,0,1,1,1,2,2,3,3,4]
#### Output: 5, nums = [0,1,2,3,4,_,_,_,_,_]

## Approach 1 - Two pointer
Solves the problem in O(n) time and O(1) space

1. If array is empty return 0
2. *if not nums:*
        *return 0*

3. Initialize insert_pos = 1 (since first element is always unique)
4. *insert_pos=1*

5. Use a pointer i starting from index 1 and iterates through all the elements in the array
6. For each element in num[i]
7. *for i in range(1,len(nums)):*

8. if nums[i] is different from nums[insert_pos - 1] (the last unique element that was placed)
9. place nums[i] at position insert_pos
10. increment insert_pos
11. Return insert_pos
12. *if nums[i] !=insert_pos[i-1]* 
13. *nums[insert_pos] = nums[i]*
14. insert_pos +=1
15. return insert_pos

```python
from typing import List

class Solution():
    def removedup(self,nums: List[int]) -> int:
        if not nums:
            return 0
        insert_pos = 1
        for i in range(1, len(nums)):
            if nums[i] != nums[i-1]:
                nums[insert_pos]= nums[i]
                insert_pos +=1
        return insert_pos        

```

## topics covered 
Array Traversal: Iterating through elements using loops.

Two Pointer Technique: Using two indices to manage positions in the array.

Conditional Statements: Checking if elements match the value to remove.

In-place Modification: Updating the array without extra space.

Edge Case Handling: Managing empty arrays.

Return Statement: Returning the count of valid elements.

Type Hinting: Using type hints for function parameters and return values.

```python
from typing import List
class Solution:
    def Searchindex(self, nums:List[int], target: int)->int:
        left, right= 0, len(nums)-1

        while left <=right:
            mid=(left+right)//2

            if nums[mid] == target:
                return mid
            elif nums[mid] < target:
                left=mid+1
            else:
                right = mid - 1
        return left            
```

```python
sol = Solution()
sol.Searchindex([1,2,3,4,5],6)
```
