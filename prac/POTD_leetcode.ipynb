# 2840. Check if Strings Can be Made Equal With Operations II

two strings are given 
s1 and s2 both of length n consisiting of lower case english letters

apply the following operations on those 2 strings any number of times

1. choose any two indices i and j 
2. where i < j and j - i is even.
3. then swap the 2 characters at those indices in the string 

Return True if you can make the strings s1 and s2 equal, and false otherwise

```python
class Solution:
    def checkStrings(self, s1: str, s2: str) -> bool:
        even1 = s1[0::2]
        odd1 = s1[1::2]
        even2 = s2[0::2]
        odd2 = s2[1::2]

        return sorted(even1) == sorted(even2) and sorted(odd1) == sorted(odd2)
```

# main idea 

the given two strings are to be arranged in such as way that the pair of odd and even elements of the indices in the string (s1 and s2) should be the same.

1. extract even-indexed characters for s1 and s2
    even1 = s1[0::2]
    even2 = s2[0::2]
2. extract odd-indexed characters for s1 and s2
    odd1 = s1[1::2]
    odd2 = s2[1::2]
3. check if even1 == even2 and odd1 == odd2

You are given two strings, str1 and str2, of lengths n and m, respectively.

A string word of length n + m - 1 is defined to be generated by str1 and str2 if it satisfies the following conditions for each index 0 <= i <= n - 1:

If str1[i] == 'T', the substring of word with size m starting at index i is equal to str2, i.e., word[i..(i + m - 1)] == str2.
If str1[i] == 'F', the substring of word with size m starting at index i is not equal to str2, i.e., word[i..(i + m - 1)] != str2.
Return the lexicographically smallest possible string that can be generated by str1 and str2. If no string can be generated, return an empty string "".

 

Example 1:

Input: str1 = "TFTF", str2 = "ab"

Output: "ababa"

Explanation:

The table below represents the string "ababa"
Index	T/F	Substring of length m
0	'T'	"ab"
1	'F'	"ba"
2	'T'	"ab"
3	'F'	"ba"
The strings "ababa" and "ababb" can be generated by str1 and str2.

Return "ababa" since it is the lexicographically smaller string.

Example 2:

Input: str1 = "TFTF", str2 = "abc"

Output: ""

Explanation:

No string that satisfies the conditions can be generated.

Example 3:

Input: str1 = "F", str2 = "d"

Output: "a"

 

Constraints:

1 <= n == str1.length <= 104
1 <= m == str2.length <= 500
str1 consists only of 'T' or 'F'.
str2 consists only of lowercase English characters.

```python
class Solution:
    def generateString(self, str1: str, str2: str) -> str:
        n, m = len(str1), len(str2)

        word = ['?']*( n+m -1)

        # Apply T constraints
        for i in range(n):
            if str1[i] == 'T':
                for j in range(m):
                    if word[i + j] == '?' or word[i + j] == str2[j]:
                        word[i + j] = str2[j]
                    else:
                        return ""
        def matches(i):
            for j in range(m):
                if word[i + j ] != str2[j]:
                    return False
            return True
        
        # Filling of the empty,[?]
        for k in range(len(word)):
            if word[k] == '?':
                for ch in 'abcdefghijklmnopqrstuvwxyz':
                    word[k] = ch

                    valid = True
                    start = max(0, k - m + 1)
                    end = min(n, k +1)

                    for i in range(start, end):
                        if str1[i]=='F':
                            fully_filled = True
                            for j in range(m):
                                if word[i + j] == '?':
                                    fully_filled = False
                                    break
                    if valid:
                        break
                    else:
                        return ""
                    
        # Step 3: Final Validation
        for i in range(n):
            if str1[i] == 'T' and not matches(i):
                return ""
            if str1[i] == 'F' and matches(i):
                return ""
        
        return ''.join(word)
```
