# 2840. Check if Strings Can be Made Equal With Operations II

two strings are given 
s1 and s2 both of length n consisiting of lower case english letters

apply the following operations on those 2 strings any number of times

1. choose any two indices i and j 
2. where i < j and j - i is even.
3. then swap the 2 characters at those indices in the string 

Return True if you can make the strings s1 and s2 equal, and false otherwise

```python
class Solution:
    def checkStrings(self, s1: str, s2: str) -> bool:
        even1 = s1[0::2]
        odd1 = s1[1::2]
        even2 = s2[0::2]
        odd2 = s2[1::2]

        return sorted(even1) == sorted(even2) and sorted(odd1) == sorted(odd2)
```

```python
sol = Solution()
sol.checkStrings('abcd','cdab')
```

# main idea 

the given two strings are to be arranged in such as way that the pair of odd and even elements of the indices in the string (s1 and s2) should be the same.

1. extract even-indexed characters for s1 and s2
    even1 = s1[0::2]
    even2 = s2[0::2]
2. extract odd-indexed characters for s1 and s2
    odd1 = s1[1::2]
    odd2 = s2[1::2]
3. check if even1 == even2 and odd1 == odd2

You are given two strings, str1 and str2, of lengths n and m, respectively.

A string word of length n + m - 1 is defined to be generated by str1 and str2 if it satisfies the following conditions for each index 0 <= i <= n - 1:

If str1[i] == 'T', the substring of word with size m starting at index i is equal to str2, i.e., word[i..(i + m - 1)] == str2.
If str1[i] == 'F', the substring of word with size m starting at index i is not equal to str2, i.e., word[i..(i + m - 1)] != str2.
Return the lexicographically smallest possible string that can be generated by str1 and str2. If no string can be generated, return an empty string "".

 

Example 1:

Input: str1 = "TFTF", str2 = "ab"

Output: "ababa"

Explanation:

The table below represents the string "ababa"
Index	T/F	Substring of length m
0	'T'	"ab"
1	'F'	"ba"
2	'T'	"ab"
3	'F'	"ba"
The strings "ababa" and "ababb" can be generated by str1 and str2.

Return "ababa" since it is the lexicographically smaller string.

Example 2:

Input: str1 = "TFTF", str2 = "abc"

Output: ""

Explanation:

No string that satisfies the conditions can be generated.

Example 3:

Input: str1 = "F", str2 = "d"

Output: "a"

 

Constraints:

1 <= n == str1.length <= 104
1 <= m == str2.length <= 500
str1 consists only of 'T' or 'F'.
str2 consists only of lowercase English characters.

```python
class Solution:
    def generateString(self, str1: str, str2: str) -> str:
        n, m = len(str1), len(str2)

        word = ['?']*( n+m -1)

        # Apply T constraints
        for i in range(n):
            if str1[i] == 'T':
                for j in range(m):
                    if word[i + j] == '?' or word[i + j] == str2[j]:
                        word[i + j] = str2[j]
                    else:
                        return ""
        def matches(i):
            for j in range(m):
                if word[i + j ] != str2[j]:
                    return False
            return True
        
        # Filling of the empty,[?]
        for k in range(len(word)):
            if word[k] == '?':
                for ch in 'abcdefghijklmnopqrstuvwxyz':
                    word[k] = ch

                    valid = True
                    start = max(0, k - m + 1)
                    end = min(n, k +1)

                    for i in range(start, end):
                        if str1[i]=='F':
                            fully_filled = True
                            for j in range(m):
                                if word[i + j] == '?':
                                    fully_filled = False
                                    break

                            if fully_filled and matches(i):
                                valid = False
                                break
                    if valid:
                        break
                else:
                    return ""
                    
        # Step 3: Final Validation
        for i in range(n):
            if str1[i] == 'T' and not matches(i):
                return ""
            if str1[i] == 'F' and matches(i):
                return ""
        
        return ''.join(word)
```

```python
sol = Solution()
print(sol.generateString('TFTF','ab'))
```

# 2751. Robot Collisions
Hard

There are n 1-indexed robots, each having a position on a line, health, and movement direction.

You are given 0-indexed integer arrays positions, healths, and a string directions (directions[i] is either 'L' for left or 'R' for right). All integers in positions are unique.

All robots start moving on the line simultaneously at the same speed in their given directions. If two robots ever share the same position while moving, they will collide.

If two robots collide, the robot with lower health is removed from the line, and the health of the other robot decreases by one. The surviving robot continues in the same direction it was going. If both robots have the same health, they are both removed from the line.

Your task is to determine the health of the robots that survive the collisions, in the same order that the robots were given, i.e. final health of robot 1 (if survived), final health of robot 2 (if survived), and so on. If there are no survivors, return an empty array.

Return an array containing the health of the remaining robots (in the order they were given in the input), after no further collisions can occur.

Note: The positions may be unsorted.

 
 

Example 1:



Input: positions = [5,4,3,2,1], healths = [2,17,9,15,10], directions = "RRRRR"
Output: [2,17,9,15,10]
Explanation: No collision occurs in this example, since all robots are moving in the same direction. So, the health of the robots in order from the first robot is returned, [2, 17, 9, 15, 10].
Example 2:



Input: positions = [3,5,2,6], healths = [10,10,15,12], directions = "RLRL"
Output: [14]
Explanation: There are 2 collisions in this example. Firstly, robot 1 and robot 2 will collide, and since both have the same health, they will be removed from the line. Next, robot 3 and robot 4 will collide and since robot 4's health is smaller, it gets removed, and robot 3's health becomes 15 - 1 = 14. Only robot 3 remains, so we return [14].
Example 3:



Input: positions = [1,2,5,6], healths = [10,10,11,11], directions = "RLRL"
Output: []
Explanation: Robot 1 and robot 2 will collide and since both have the same health, they are both removed. Robot 3 and 4 will collide and since both have the same health, they are both removed. So, we return an empty array, [].
 

Constraints:

1 <= positions.length == healths.length == directions.length == n <= 105
1 <= positions[i], healths[i] <= 109
directions[i] == 'L' or directions[i] == 'R'
All values in positions are distinct

1. create a list of robots with:
    (position, health, direction, original_idex)

2. sort robots by position

3. Initialize empty stack

4. Traverse robots:
        if direction =='R':
            push robot into stack
        
        else (direction == 'L'):
            while stack not empty AND current robot alive:
                compare with top of stack

                if stack robot health < current:
                    pop stack 
                    current health -=1
                
                elif stack robot health > current:
                    stack robot health -1

                else:
                    pop stack
                    current robot dies

            if current robot still alive:
                add it to survivors

5. add remaining stack robots to survivors 

6. sort survivors by original index

7. Return their healths


```python
class Solution:
    def survivedRobotsHealths(self, positions, healths, directions):
        n=len(positions)

        robots=[]
        for i in range(n):
            robots.append((positions[i],healths[i],directions[i], i))

        robots.sort()

        stack = []
        survivors = []

        for pos, health, direction, idx in robots:

            if direction == 'R':
                stack.append([pos,health,direction,idx])
            else:
                curr_health = health 

                while stack and curr_health > 0:
                    last = stack[-1]

                    if last[1] < curr_health:
                        stack.pop()
                        curr_health -=1

                    elif last[1] > curr_health:
                        last[1] -=1
                        curr_health = 0 
                    else:
                        stack.pop()
                        curr_health = 0
                
                if curr_health > 0:
                    survivors.append([pos, curr_health, direction, idx])
        
        survivors.extend(stack)

        survivors.sort(key=lambda x: x[3])

        return [robot[1] for robot in survivors]

        
```
